```python
"""
Використовуючи дані із файлу cities.csv створіть список типу:
cities = [['city1', 'city2', km]...]. З отриманого списку створить граф.
Візуалізуйте отриманий граф.
"""
```

```python
import csv
```

```python
with open('cities.csv', 'r') as file:
    reader = csv.reader(file, delimiter=';')
    cities = []
    for row in reader:
        cities.append(row)
```

```python
print(cities)
```

```python
import networkx as nx
```

```python
# Створюємо порожній граф
graph = nx.Graph()
```

```python
# Додаємо вузли (міста) до графа
for city1, city2, distance in cities:
    graph.add_node(city1)
    graph.add_node(city2)
```

```python
# Додаємо ребра (дороги) до графа
for city1, city2, distance in cities:
    graph.add_edge(city1, city2, weight=int(distance))
```

```python
# Виводимо інформацію графа
print(f"Number of nodes: {graph.number_of_nodes()}")
print(f"Number of edges: {graph.number_of_edges()}")
print(f"Node List: {list(graph.nodes)}")
print(f"Edges List: {list(graph.edges)}")
```

```python
import matplotlib.pyplot as plt
```

```python
# Візуалізуємо граф
pos = nx.spring_layout(graph, seed=42)
nx.draw_networkx_nodes(graph, pos, node_color='lightblue')
nx.draw_networkx_labels(graph, pos, font_size=10, font_family='Arial')
nx.draw_networkx_edges(graph, pos, edge_color='gray', width=1, alpha=0.7)
nx.draw_networkx_edge_labels(graph, pos, font_size=8, edge_labels=nx.get_edge_attributes(graph, 'weight'))
plt.axis('off')
plt.show()
```

```python
"""
Напишіть функцію знаходження найкоротшого маршруту між двома
населеними пунктами, яка приймає у якості аргументів об'єкт
графу і пару населених пунктів, а повертає маршрут і його
протяжність.
"""
```

```python
def shortest_path(graph, start_city, end_city):
    # Знаходимо найкоротший шлях між початковим та кінцевим містами
    path = nx.shortest_path(graph, start_city, end_city, weight='weight')
    # Знаходимо протяжність маршруту
    distance = nx.shortest_path_length(graph, start_city, end_city, weight='weight')
    # Повертаємо маршрут та його протяжність
    return path, distance
```
